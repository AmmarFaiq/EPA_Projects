This is just a quick example of generating Dirichlet distributions and plotting them. 

```python
import ternary
import random
import numpy as np
```

```python

def random_Dirichlet(num_points=25, scale=100):
    points = []
    for i in range(num_points):
        x = np.random.gamma(4.23,1)
        y = np.random.gamma(2.32,1)
        z = np.random.gamma(83.71,1)
        xx = x*100/(x+y+z)
        yy = y*100/(x+y+z)
        zz = z*100/(x+y+z)
        points.append((xx,yy,zz))
    return points
```

```python

# Scatter Plot of Real Data
scale = 100
figure, tax = ternary.figure(scale=scale)
figure.set_size_inches(10, 10)
tax.set_title("Random Points", fontsize=20)
tax.boundary(linewidth=2.0)
tax.gridlines(multiple=10, color="blue")
# Plot a few different styles with a legend
rnd_points = random_Dirichlet(160, scale=scale)
tax.scatter(rnd_points, marker='s', color='red', label="Red Squares")

tax.ticks(axis='lbr', linewidth=1, multiple=20)
tax.clear_matplotlib_ticks()

tax.show()
```
