### Getting shapefiles to plot districts

```python
import shapefile
sh = shapefile.Reader('C:/Users/Михаил/AnacondaProjects/Disinofrmation project/Wijkgrenzen_Den_Haag/Wijkgrenzen_Den_Haag')
```

```python
# Processing shapefiles
# Gettign shapefile records 
sh_recs = sh.records()

# Lists of the names to store particular values
district_names = list()

# Getting the names
for i in range(len(sh_recs)):
    district_names.append(sh_recs[i][2])

# Get the circle coordinates as example
district_coors = [s.points for s in sh.shapes()]
test_lon = np.empty(len(district_coors[0]))
test_lat = np.empty(len(district_coors[0]))

lons_combined = np.empty((44, 0)).tolist()
lats_combined = np.empty((44, 0)).tolist() 
for i in range(len(district_coors)):
    for j in range(len(district_coors[i])):
        lons_combined[i].append(district_coors[i][j][0])
        lats_combined[i].append(district_coors[i][j][1])
```

### Plotting interesting data

```python
from bokeh.io import show
from bokeh.models import (ColumnDataSource, HoverTool, LogColorMapper, LinearColorMapper)
from bokeh.palettes import Viridis6 as palette
from bokeh.plotting import figure

palette = ['#CC3300', '#CF3C09',  '#D14411',  '#D44D1A',  '#D65523', '#D95E2B', '#DB6634',  
          '#DE6F3D',  '#E07745',  '#E3804E',  '#E68856',  '#E8915F', '#EB9A68',  '#EDA270', 
          '#F0AB79',  '#F2B382',  '#F5BC8A',  '#F7C493',  '#FACD9C',  '#FCD5A4',  '#FFDEAD']

palette.reverse()

color_mapper = LinearColorMapper(palette=palette)

source = ColumnDataSource(data=dict(
    x = lons_combined,
    y = lats_combined,
    name = district_names,
    population = list_to_plot,
))

TOOLS = "pan, wheel_zoom, reset, hover, save"

p = figure(
    title = "Den Haag, population per district", tools = TOOLS,
    x_axis_location = None, y_axis_location = None
)

p.grid.grid_line_color = None

p.patches('x', 'y', source=source,
          fill_color={'field': 'population', 'transform': color_mapper},
          fill_alpha=0.9, line_color="white", line_width=0.5)

hover = p.select_one(HoverTool)
hover.point_policy = "follow_mouse"
hover.tooltips = [
    ("Name", "@name"),
    ("Person in district", "@population")
#     ("(Long, Lat)", "($x, $y)"),
]

show(p)
```
