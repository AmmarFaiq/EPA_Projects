# Load data file

```python
# path of csv files
# Municipal election data (2014)
vote2014 ='D:\\Jupyter_File\\Hackaton_Friday\\Dataset_haag\\voting_2014.csv'
# House of representatives election data (2017)
vote2017 = 'D:\\Jupyter_File\\Hackaton_Friday\\Dataset_haag\\House of Representatives 2017 - Districts.csv'

# convert csv to dataframe
df_ori_vote2014  = pd.read_csv(vote2014,index_col=None, header=0)
# df_ori_vote2014['Wijken'] = df_ori_vote2014['Wijken'].apply(lambda x: x[3:])
df_ori_vote2014 = df_ori_vote2014.set_index('Wijken')
df_ori_vote2017  = pd.read_csv(vote2017,index_col=None, header=0)
# df_ori_vote2017['Wijken'] = df_ori_vote2017['Wijken'].apply(lambda x: x[3:])
df_ori_vote2017 = df_ori_vote2017.set_index('Wijken')

# summing values of each districts
df_ori_vote2014['Total'] =  df_ori_vote2014.apply(lambda x: x.sum(), axis = 1)
df_ori_vote2017['Total'] =  df_ori_vote2017.apply(lambda x: x.sum(), axis = 1)

# fill the NaN values
df_vote2014 = df_ori_vote2014.fillna(0)
df_vote2017 = df_ori_vote2017.fillna(0)
df_ori_vote2014 = df_ori_vote2014.fillna(0)
df_ori_vote2017 = df_ori_vote2017.fillna(0)

# df_vote2014 = df_vote2014[(df_vote2014.T != 0).any()]
# df_vote2017 = df_vote2017[(df_vote2017.T != 0).any()]

# df_ori_vote2014 = df_ori_vote2014[(df_ori_vote2014.T != 0).any()]
# df_ori_vote2017 = df_ori_vote2017[(df_ori_vote2017.T != 0).any()]

# Check the quantity of the data
print('2014 Voting Data consists :',len(df_vote2014))
print('2017 Voting Data consists :',len(df_vote2017))
print('2014 Political Parties :',len(df_vote2014.columns)-1)
print('2017 Political Parties :',len(df_vote2017.columns)-1)
```

```python
# normalize the voting based on its value divided by total voting in that districts
df_ori_vote2014_normalize = df_vote2014.loc[:,'PvdA':'TerLin'].div(df_vote2014['Total'], axis=0).fillna(0)
df_ori_vote2017_normalize = df_vote2017.loc[:,' VVD':' VDP'].div(df_vote2017['Total'], axis=0).fillna(0)
```

#### Visualize the initial voting data

```python
# 2014 municipal election
plt.figure(figsize = (13,7))
sns.heatmap(df_ori_vote2014_normalize)
```

# 2. NMF 2014 municipal election data NMF

```python
# Fit dataframe into NMF 

N_com = 7

model_2014 = NMF(n_components = N_com)
W_vote_2014 = model_2014.fit_transform(df_ori_vote2014_normalize)
H_vote_2014 = model_2014.components_

# Record the districts and variables names
district_name_2014 = df_ori_vote2014_normalize.index
variable_name_2014 = df_ori_vote2014_normalize.columns

NMF_vote_W_2014 = pd.DataFrame(index=district_name_2014, columns= list(range(0, N_com)))
NMF_vote_H_2014 = pd.DataFrame(index= list(range(0, N_com)) , columns= variable_name_2014)

# fit the dataframe with its coresponding name
NMF_vote_W_2014.loc[:,:] = W_vote_2014
NMF_vote_H_2014.loc[:,:] = H_vote_2014
```

```python
NMF_vote_W_2014.head()
```

## 2.1 Normalized with the sum NMF of district

```python
NMF_vote_W_2014['Total'] =  NMF_vote_W_2014.apply(lambda x: x.sum(), axis = 1)
NMF_vote_W_2014_normalize = NMF_vote_W_2014.iloc[:,0:7].div(NMF_vote_W_2014.Total, axis=0) .fillna(0)
NMF_vote_W_2014_normalize.head()
```

```python
plt.figure(figsize = (13,7))
sns.heatmap(NMF_vote_W_2014_normalize, robust=True ,annot = True)
```

```python
plt.figure(figsize = (13,7))
sns.clustermap(NMF_vote_W_2014_normalize.T, annot = True)
```
