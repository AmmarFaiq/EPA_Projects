# Parse the Data

```python
filename = "DemoVoteData/uitslagen2017.csv"
```

## Create a List of Lists

```python
import csv
data = []
gemeente = []
with open(filename, 'r') as csvfile:
    votedata = csv.reader(csvfile, delimiter=';', quotechar='"')

    
    for n,row in enumerate(votedata):

        if (n==0):
            header = [row]
        else:
            data_row = []
            gemeente_row = []
            for i in range(0,2):
                gemeente_row.append(row[i])
            for i in range(3,32):
                if (row[i]==''):
                    n=0
                else:
                    n=int(row[i])
                data_row.append(n)
        gemeente.append(gemeente_row)
        data.append(data_row)
```
