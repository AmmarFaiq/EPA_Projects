```python
from sklearn.decomposition import NMF

N_com = 5

model = NMF(n_components = N_com, init='random', random_state=0)
W = model.fit_transform(district)
H = model.components_
```

```python
district_name = district.index
variable_name = district.columns

NMF_W = pd.DataFrame(index=district_name, columns= list(range(0, N_com)))
NMF_H = pd.DataFrame(index= list(range(0, N_com)) , columns= variable_name)
```

```python
NMF_W.loc[:,:] = W
NMF_H.loc[:,:] = H
```

```python
plt.figure(figsize = (16,8))
sns.heatmap(NMF_W)
```

```python
plt.figure(figsize = (16,8))
sns.clustermap(NMF_W.T)
```

```python
plt.figure(figsize = (16,8))
sns.heatmap(NMF_H)
```

```python
plt.figure(figsize = (16,8))
sns.clustermap(NMF_H)
```
