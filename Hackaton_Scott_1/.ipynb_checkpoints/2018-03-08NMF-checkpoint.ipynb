### Uploading necessary packages

```python
import numpy as np
import pandas as pd
import os
 
from sklearn.preprocessing import MaxAbsScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing the data 
df = pd.read_excel('Data.xls')
df.head()
```

```python
# The total population of a particular district
delimiter = df['Number of employed persons']

# Normalizing occupation data
for i in range(2, 10):
    df[df.columns[i]] = df[df.columns[i]] / delimiter
# df.head()

# Normalizing housing and non-housing density values
for i in range(10, 12):
    df[df.columns[i]] = df[df.columns[i]] / 100
# df.iloc[0:5,10:12]

# Normalizing households data
delimiter = df['Number of households']
for i in range(13, 17):
    df[df.columns[i]] = df[df.columns[i]] / delimiter
# df.iloc[0:5,13:18]

# Normalizing population data
delimiter = df['Total population of the district']
for i in range(18, 29):
    df[df.columns[i]] = df[df.columns[i]] / delimiter
# df.iloc[0:5,17:29]

# Normalizing crimes data
delimiter = df['Total population of the district']
df[df.columns[0]] = df[df.columns[0]] / delimiter
# df.iloc[0:5, 0]
```

```python
# Dropping no more important columns 
df = df.drop(['Number of employed persons', 'Number of households', 'Total population of the district'], axis = 1)

# Filling NaNs with zeors
df.fillna(0)
```

```python
# Saving the result
# Due to some reasons we need manually fill NaNs
df.to_excel('NormalizedDataSet.xlsx')
```

```python
# Loading data set after NaNs were filled by zeros
df = pd.read_excel('NormalizedDataSet.xlsx')
district_names = df.index
factor_names =  df.columns
```

```python
# Normalizing the whole df
df = MaxAbsScaler().fit_transform(df)
```

### Applying NMF

```python
from sklearn.decomposition import NMF

model = NMF(n_components = 4)
model.fit(df)
nmf_features = model.transform(df)
components = model.components_

W = nmf_features
H = components

W_df = pd.DataFrame(index = district_names, columns = list(range(0, 4)))
H_df = pd.DataFrame(index = list(range(0, 4)) , columns = factor_names)

W_df.loc[:,:] = W
H_df.loc[:,:] = H
```

```python
plt.figure(figsize = (16,8))
sns.clustermap(W_df, annot = True)
plt.show()
```

```python
plt.figure(figsize = (16,8))
sns.clustermap(H_df.T, annot = True)
plt.show()
```

### Applying PCA

```python
from sklearn.decomposition import PCA
from scipy.stats import pearsonr
```

```python
# Functions to highlight the most interesting values
def highlight_max(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_max = s >= 0.5
    return ['background-color: yellow' if v else '' for v in is_max]

def highlight_min(s):
    '''
    highlight the maximum in a Series yellow.
    '''
    is_min = s <= -0.5
    return ['background-color: yellow' if v else '' for v in is_min]
```

```python
# Checking the correlation of the factors 
correlation_table = df.corr(method='pearson')
correlation_table.style.apply(highlight_min)
```

```python
# Perform the necessary imports
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt

# Create scaler: scaler
scaler = StandardScaler()

# Create a PCA instance: pca
pca = PCA()

# Create pipeline: pipeline
pipeline = make_pipeline(scaler, pca)

# Fit the pipeline to df
pipeline.fit(df)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.show()
```
